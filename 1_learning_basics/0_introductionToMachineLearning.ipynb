# Importance of spliting our data

Model Validation \
you will want to evaluate every model you build, to check the quality of the model. \
there are many ways to do this, but we will use MAE (Mean Absolute Error) \
error = actual - predicted (we take absolute values of all the error) \
we take mean of all the errors and that is called the MAE

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

what we did is called a `insample score`, we used the same data for building the model and also evaluating the model. \

imagine we took `door colour` as part of out features. our model during evaluation might have a pattern of high cost for a certain colour, but the pattern might not hold for new data. hence, the model will be inaccurate when we predict with it. \

the fix is simple, we need data we didn't use to build the model to evaluate the model.

we can use `train_test_split` built in the scikit-learn package.

```python
from sklearn.model_selection import train_test_split
```

spliting the data is based on a random number genrator. hence, it also have a seed / random_state

```python
train_X,val_X,train_y,val_y = train_test_split(X,y,random_state = 0)
```

```python
melbourne_model = DecisionTreeRegressor(random_state=0)
melbourne_model.fit(train_X,train_y)
```

```python
val_predict = melbourne_model.predict(val_X)
print("MAE is:",mean_absolute_error(val_y,val_predict))
```

our MAE for `in-sample` data was 1115 dollars and for testing data it is 274,037 dollars 

now the difference of predicted and actual values is large.

# Underfitting and overfitting

we will tweak the depth
no of nodes per split = 2**n n= split number 

overfitting
divde a lot -> less house per leaf -> predict close to those leaf -> bad prediction
this does very good on training data but bad on validation data

underfitting
divde to less -> to much variety -> bad prediction
![alt text](image.png)

```python
def get_mae(max_leaf_nodes,train_X,val_X,train_y,val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=0)
    model.fit(val_X,val_y)
    pred_val = model.predict(val_X)
    mae = mean_absolute_error(pred_val,val_y)
    return mae
```

```python
for max_leaf_nodes in [5,50,500,5000,50000]:
    my_mae = get_mae(max_leaf_nodes,train_X,val_X,train_y,val_y)
    print("max leaf nodes: %d \t\t MAE: %d"%(max_leaf_nodes,my_mae))
```
