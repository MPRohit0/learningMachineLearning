we can also search based on conditions

```python
dataset['Rooms'] == 3 #dataset.Rooms == 3
```

```python
dataset.loc[dataset.Rooms==3]
```

```python
# we can also chain them
dataset.loc[(dataset.Rooms==3)&(dataset.Distance<7)]
```

some inbuilt feature:
1. dataset.Rooms.isin([2,3]) is just dataset.loc[(dataset.Rooms == 2) | (dataset.Rooms == 3)]

```python
dataset.loc[(dataset.Rooms == 2) | (dataset.Rooms == 3)]
```

```python
dataset.loc[dataset.Price.notnull()]
```

# maping a fuction

lambda function - lambda x: x+10

```python
dataset.Rooms.map(lambda x:x-1)
```

```python
def remean_points(row):
    row.Car = row.Car + 1
    return row

dataset.apply(remean_points,axis='columns') #if axis = index, we will get each column
```

```python
dataset.Rooms.head().map(lambda x:str(x)) +"-" +dataset.Car.head().map(lambda x:str(int(x)))
```
